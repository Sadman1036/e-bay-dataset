#Renaming and lowercasing Column names

```python
def rename_coulmns(col):
    col = col.replace('registraion year','registration_year')
    col = col.replace('registration __month','registration_month')
    col = col.replace('unrepaired','notRepairedDamage')
    col = col.replace('datecreated','date_created')
    col = re.sub( '(?<!^)(?=[A-Z])', '_', col ).lower()
    return col
```

```python
df.columns = [rename_coulmns(cols) for cols in df.columns]
df.columns
```

```python
top_brands= df["brand"].value_counts(normalize=True).round(3)
most_common_brands = top_brands[top_brands>0.05].index
most_common_brands
```

We can see that there is a price gap between the top 5 brands in the sales data. We can see that cars manufactured by Audi, BMW and Mercedes Benz tend to be priced higher than the competition. Opel is the least expensive of the top 5 brands while Volkswagen is in between. This could be one of the reasons for the popularity of Volkswagen cars.

```python
def all_means(most_common_brands):
    brand_mean_price = {}

    for brands in most_common_brands:
        brands_only = df[df["brand"]== brands]
        brand_meanPrice = brands_only["price"].mean()
        brand_mean_price[brands] = int(brand_meanPrice)

    brand_mean_mileage = {}

    for brands in most_common_brands:
        brands_only = df[df["brand"]==brands]
        brand_mileage = brands_only["odometer_km"].mean()
        brand_mean_mileage[brands] =  int(brand_mileage)
   
    return brand_mean_price,brand_mean_mileage
```
